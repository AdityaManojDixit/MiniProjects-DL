```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

print(sigmoid(10))
```

```python
def calcTan(x):
  return math.tan(math.radians(x))

print(calcTan(45))
```

```python
x_values = np.linspace(-90, 90, 1000) #Generating array of 1000 values (angles)

y_values = []

for i in x_values:
  tanx = calcTan(i)
  y_values.append(tanx)
```

```python
plt.plot(x_values, y_values, label='tan(x)')
plt.xlabel('Angle (degrees)')
plt.ylabel('Tangent')
plt.title('Tangent Function')
plt.show()
```
